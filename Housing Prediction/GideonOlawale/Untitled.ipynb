```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

transformer = ColumnTransformer([('scaler',StandardScaler(),[0,1,2]),
                                ('category',OneHotEncoder(sparse = False),[3,4,5,6,7,8])],
                                remainder = 'passthrough')
```

```python
scaler = StandardScaler()
yt = scaler.fit_transform(y)
```

```python
xt = transformer.fit_transform(dataset5)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import (LinearRegression, LogisticRegression, ElasticNet, Lasso, Ridge,HuberRegressor)
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
import time as t
estimators = {
    'linear': LinearRegression(),
    'decision_tree': DecisionTreeRegressor(),
    'elastic': ElasticNet(),
    'lasso' : Lasso(),
    'ridge' : Ridge(),
    
    'knn'   : KNeighborsRegressor()
}

for name,estimator in estimators.items():
    t1 = t.time()
    kfold = KFold(n_splits = 10, shuffle = True, random_state = 100)
    score = cross_val_score(estimator = estimator, X = xt, y = yt, cv = kfold)
    print(f'{name:<10},Average R value for the 10 iterations: {score.mean():<8.4f}Standard Deviation {score.std():.3f}:Time to run this estimator is {t.time() - t1:.3f}Seconds')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(dataset5, y, random_state = 100, test_size = 0.1)
```

```python
len(y_test)
```

```python
from tempfile import mkdtemp
from shutil import rmtree
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

cachedir = mkdtemp()
parameters = {  'min_samples_leaf' : range(1,4),
                 'max_depth' : range(1,15),
                  'min_samples_split' : range(5,20,2)
              }

pipe = Pipeline([('trans',transformer),
                 ('grid',GridSearchCV(estimator= DecisionTreeRegressor(), param_grid = parameters, n_jobs= -1, cv = KFold(10,shuffle = True, random_state = 100)))],
                  memory = cachedir)
rmtree(cachedir)
```

```python
pipe.fit(X_train,y_train)
```

```python
pipe.named_steps['grid'].best_params_
```

```python
pipe.named_steps['grid'].best_score_
```

```python
pipe.named_steps
```

```python
prediction = pipe.predict(X_test)
```
